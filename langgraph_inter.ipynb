```python
from langgraph.graph import StateGraph, END
from typing import TypedDict, Optional
```

```python
class GraphState(TypedDict):
    user_input: str
    documents: Optional[str]
    is_valid: Optional[bool]
    answer: Optional[str]
    retry_count: int
```

```python
def ingest_node(state: GraphState):
    print("Ingesting document... attempt:", state["retry_count"])

    if state["retry_count"] < 1:
        fake_doc = ""
    else:
        fake_doc = "LangGraph is used to build stateful AI workflows."

    return {
        "documents": fake_doc,
        "retry_count": state["retry_count"] + 1
    }
```

```python
def confirm_node(state):
    print("Confirming document validity...")

    docs = state.get("documents", "")

    is_valid = docs is not None and len(docs.strip()) > 0

    print("Valid?" , is_valid)

    return {
        "is_valid": is_valid
    }
```

```python
def respond_node(state: GraphState):
    print("Generating response...")

    user_q = state["user_input"]
    docs = state["documents"]

    response = f"Based on the document, the answer to '{user_q}' is:\n\n{docs}"

    return {
        "answer": response
    }
```

```python
graph = StateGraph(GraphState)

graph.add_node("ingest", ingest_node)
graph.add_node("confirm", confirm_node)
graph.add_node("respond", respond_node)
```

```python
def decide_next(state: GraphState):
    if state["is_valid"]:
        print("Document valid ‚Üí going to respond")
        return "respond"
    else:
        print("Invalid ‚Üí retrying ingestion")
        return "ingest"
```

```python
graph.set_entry_point("ingest")

graph.add_edge("ingest", "confirm")
graph.add_conditional_edges(
    "confirm",
    decide_next,
    ["respond", "ingest"]
)
graph.add_edge("respond", END)
```

```python
app = graph.compile()
```

```python
from IPython.display import Image, display

display(
    Image(
        app.get_graph().draw_mermaid_png()
    )
)
```

```python
initial_state = {
    "user_input": "What is LangGraph?",
    "documents": None,
    "is_valid": None,
    "answer": None,
    "retry_count": 0
}

result = app.invoke(initial_state)

print("\nFINAL OUTPUT\n")
print(result["answer"])
```
